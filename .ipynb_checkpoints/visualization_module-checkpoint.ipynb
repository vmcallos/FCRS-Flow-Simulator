```python
import pandas as pd
%matplotlib inline
```

```python
#import data frame for model results;
df_out = pd.read_excel('model_results.xls', sheetname=0)
```

```python
'''Add new calculated fields'''

df_out['dSF']=df_out['SF_in']-df_out['SF_out']
df_out['d%SF']=100*df_out['dSF']/df_out['SF_in']
df_out['dP']=df_out['InletPressure']-df_out['OutletPressure']
df_out['sum'] = df_out['HeatLoss'] + df_out['SkinFriction'] + df_out['Expansion'] + df_out['Fittings'] + df_out['K.E.']
df_out['%HeatLoss'] = 100* df_out['HeatLoss'] / df_out['sum']
df_out['%SkinFriction'] = 100* df_out['SkinFriction'] / df_out['sum']
df_out['%Expansion'] = 100* df_out['Expansion'] / df_out['sum']
df_out['%Fittings'] = 100* df_out['Fittings'] / df_out['sum']
df_out['%K.E.'] = 100* df_out['K.E.'] / df_out['sum']
```

```python
df_out.iloc[:,[4,5,6,7,8, 13]].plot.bar(stacked=True, figsize=(10,6), grid=True)
```

```python
df_out.iloc[:,[21, 22,23,24, 25]].plot.bar(stacked=True, figsize=(14,7), grid=True, ylim=(0,150), \
                                           title='Percent Contribution')
```
